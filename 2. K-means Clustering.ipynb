### Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
```

### Load the datasets

```python
iris = pd.read_csv('./Iris.csv')
iris['Species'] = iris.Species.str.replace('Iris-' , '')
iris
```

### Substract the values and transform the dataset into a numpy array

```python
x = iris.iloc[:, [1,2,3,4]]
x  = np.array(x)
x
```

### Find the optimal number of clusters

```python
# Collecting the distortions into list
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(x)
    distortions.append(kmeanModel.inertia_)
```

As k (number of clusters) becomes larger, the sum of variation (distortion) within clusters becomes smaller, so how to terminate the value of k? In the following plot, when we changed the number of clusters from 1 to 3, the sum of variation reduces very sharply, we can 3 as the optimal k.

```python
# Plotting the distortions
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal clusters')
plt.show()
```

```python
# Define the model
kmeans_model = KMeans(n_clusters=3, n_jobs=3, random_state=32932)
# Fit into our dataset fit
kmeans_predict = kmeans_model.fit_predict(x)
```

```python
iris['Cluster'] = kmeans_predict
iris
```
